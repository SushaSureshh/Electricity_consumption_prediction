```python
# Read the csv and convert it into a dataframe

data = pd.read_csv('input.csv')
data.describe()
```

```python
# Extract the prediction

Y = data['KWH']
del data['KWH']
```

```python
""" I am deleting the columns with string values - 

under the assumption that they do not contribute much to the prediction task"""

del data['UR']
del data['METROMICRO']
del data['IECC_Climate_Pub']
```

```python
# Select only dtype = float, boolean or int 

data = data.select_dtypes(exclude=['object'])
```

```python
'''Normalize the integer values for better results using min_max_scaler'''

min_max_scaler = MinMaxScaler() 
column_names_to_normalize = [i for i in data.columns if 'TYPE' not in i]
x = data[column_names_to_normalize].values
x_scaled = min_max_scaler.fit_transform(x)
df_temp = pd.DataFrame(x_scaled, columns=column_names_to_normalize, index = data.index)
data[column_names_to_normalize] = df_temp
```

```python
x = Y.values
x = x.reshape(-1,1)
x_scaled = min_max_scaler.fit_transform(x)
Y1 = pd.DataFrame(x_scaled, columns=['KWH'], index = Y.index)
```

```python
# Define models RandomForest and xgboost

model1 = RandomForestRegressor(n_estimators=20, random_state=0)
model = XGBRegressor(objective='reg:squarederror',max_depth=10)
```

```python
# Split the data frame to test set and training set

X_train,X_test,Y_train,Y_test = train_test_split(data,Y,test_size=0.33, random_state=42)
```

```python
# Print the values of the Y

#Y_train[0:20]
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
params = {
        'min_child_weight': [1, 5, 10],
        'gamma': [0.5, 1, 1.5, 2, 5],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [3, 4, 5]
        }

folds = 3
param_comb = 5

skf = StratifiedKFold(n_splits=folds, shuffle = True, random_state = 1001)

random_search = RandomizedSearchCV(model, param_distributions=params, n_iter=param_comb, scoring='neg_mean_squared_error', n_jobs=4, cv=skf.split(data,Y), verbose=3, random_state=1001 )
random_search.fit(data, Y)

```

```python
print('\n Best hyperparameters:')
print(random_search.best_params_)
```

```python
# Fit model to the training data

model.fit(X_train, Y_train)
```

```python
# Predict the values using text data

y_pred = model.predict(X_test)
```

```python
# Compute the root mean squared error between the predicted value and actual value
print("\nRoot Mean Squared error:", mean_squared_error(np.round(y_pred),Y_test))
```
